# Filter Crew_in

```
m = pd.read_csv("/content/drive/MyDrive/CIS550/movies_filtered.csv")
```

```
mids = m['tconst']
mids
```

```
p = pd.read_csv("/content/drive/MyDrive/CIS550/people_filtered.csv")
```

```
pids = p['nconst']
pids
```

```
csv_file_path = '/content/drive/MyDrive/CIS550/crew_in.csv'  # Replace with the path to your CSV file

# Chunk size for reading the CSV file
chunk_size = 10**6  # Adjust the chunk size based on your available memory

# Initialize an empty DataFrame to store the filtered results
filtered_df = pd.DataFrame()

# Read the CSV file in chunks
chunks = pd.read_csv(csv_file_path, chunksize=chunk_size)

# Iterate over chunks
for chunk in chunks:
    # Filter rows based on the list of IDs
    filtered_chunk = chunk[chunk['tconst'].isin(mids) & chunk['nconst'].isin(pids)]  # Assuming 'ID' is the column containing IDs

    # Append the filtered chunk to the results DataFrame
    filtered_df = pd.concat([filtered_df, filtered_chunk], ignore_index=True)

# Display the filtered DataFrame
filtered_df.head()
```

```
filtered_df
```

```
c = filtered_df.drop(['ordering', 'job'], axis=1)
c
```

```
c['characters'] = c['characters'].astype(str).str.replace(r'\[|\]|"|\'', '')
c
```

```
filtered_df['category'].unique()
```

```
allowed_categories = ['director', 'producer', 'actor', 'actress', 'writer']

# Filter the DataFrame
c = c[c['category'].isin(allowed_categories)]
c
```

```
csv_file_path = 'crew_in_filtered.csv'  # Replace with the desired output file path
c.to_csv(csv_file_path, index=False)
```
